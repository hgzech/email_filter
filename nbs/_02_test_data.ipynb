```python
import os
from datetime import datetime
from email_filter.exchange import EmailClient, extract_domains
```

```python
USERNAME = os.environ.get("EXCHANGE_USER")
PASSWORD = os.environ.get("EXCHANGE_PASSWORD")
email_client = EmailClient("msx.tu-dresden.de", 993, USERNAME, PASSWORD)
```

### Trusted email addresses
We build the trusted addresses based on addresses sent to in the last year. We will use this as the starting point for our trusted addresses file. This file will later be updated by the system.

```python
%%time
# Fetch emails since a specific date
since_date = datetime(2022, 9, 21)  # Replace with actual date
good_emails = email_client.get_emails(from_folder = "Sent Items", since_date=since_date)
```

```python
trusted_addresses = [e.recipient + e.cc + e.bcc for e in good_emails]
trusted_addresses = list(set([a for adds in trusted_addresses for a in adds]))
```
